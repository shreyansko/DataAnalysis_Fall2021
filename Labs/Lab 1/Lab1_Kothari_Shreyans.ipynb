### 1) Choose one variable, look at its distribution, and plot a histogram of it. Explain what you take away from looking at the variable.

```python
happy_tab = pd.crosstab(index =d["happy"], columns = "count")
happy_tab #Cross tab
```

```python
#Creating a function to calculate percent
def compute_percentage(x):
    pct = (x/happy_tab['count'].sum()) * 100
    return round(pct, 2)
```

```python
# Adding percentage column to the data frame 
happy_tab['percentage'] = happy_tab.apply(compute_percentage, axis = 1)
happy_tab
```

```python
#Plotting happiness on a histogram
%matplotlib inline
d.hist(column='happy')
```

```python
happy_race = pd.crosstab(d.hhrace, d.happy )
happy_race.astype('float').div(happy_race.sum(axis=0), axis=1).round(4)*100
```

```python
#cross tabulation of self-reported happiness and household race
pd.crosstab(d.happy, d.hhrace )
```

```python
race_happy = pd.crosstab(d.happy, d.hhrace )
race_happy.astype('float').div(race_happy.sum(axis=0), axis=1).round(4)*100
```

#### Conclusions:
The happy_race cross tabulation of self-reported happiness and household race gives us an idea of the distribution of the three classes of self-reported hapiness among different household races in the study sample. Of all the respondents who reported being 'Very happy', 80.76% were white whereas only 0.22% were American Indian. Similarly, of all the respondents who reported being 'Not too happy', 70.51% were white as well; the smallest group was American Indian again at 0.26%. 

Looking at intra-race happiness levels (race_happy cross tabulation), of all: 
1. White respondents interviewed, most (56.35%) reported being 'Very Happy', and least (11.79%) reported being 'Not too happy'.
2. Black respondents interviewed, most (54.61%) reported being 'Pretty Happy', and least (27.93%) reported being 'Very  happy'.
3. American Indian respondents interviewed, most (72.73%) reported being 'Pretty Happy', and least (9.09%) reported being 'Not too  happy'.
4. Asiatic, Oriental respondents interviewed, most (58.33%) reported being 'Pretty Happy', and least (6.94%) reported being 'Not too  happy'.
6. Other, mixed respondents interviewed, most (53.89%) reported being 'Pretty Happy', and least (22.75%) reported being 'Not too  happy'.

More than a true depiction of self-reported happiness levels among different races, these results just show a highly skewed result due to a high number of white respondents, and not enough respondents from other races in the sample. The sample size being smaller than 30 for most races, the result is not statistically significant.
